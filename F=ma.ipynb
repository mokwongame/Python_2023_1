```python
import matplotlib.pyplot as plt
import numpy as np
from math import *
```

```python
a = 10
m = 50
x0 = 0
v0 = a
dt = 0.001
x1 = v0*dt
nmax = int(3*a/dt)
```

```python
def force(t):
    if t < 0: return 0
    elif t <= 2*a: return m*a*cos(2*pi/a*t)
    else: return 0
```

```python
n = range(2, nmax+1)
xl = [x0, x1]
tl = [0, dt]
for nv in n:
    xv = 2*xl[nv-1]-xl[nv-2]+dt*dt/m*force((nv-1)*dt)
    xl.append(xv) # list of x
    tl.append(nv*dt) # list of t
```

```python
plt.plot(tl, xl)
plt.grid(True)
```
