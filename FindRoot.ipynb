```python
import matplotlib.pyplot as plt
import numpy as np
from math import *
```

```python
def myExp(x):
    return exp(x) + x*x - 23
```

```python
exp(10)
```

```python
x = np.arange(2., 4., 0.001)
y = []
for xv in x:
    y.append(myExp(xv))
plt.plot(x, y)
plt.grid(True)
```
