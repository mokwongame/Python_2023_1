```python
import matplotlib.pyplot as plt
import numpy as np
from math import *
```

```python
def F(t):
    a = 12
    if t >= 0 and t <= a: return a*exp(-t) + cos(t)
    else: return 0
```

```python
F(20)
```

```python
x=list(np.arange(-5, 20, 0.01))
y=[]
for xv in x:
    y.append(F(xv))
```

```python
dt = 1e-3
m = 50
a = 12
v0 = a
x = list(np.arange(0, 2*a, dt))
```

```python
y=[]
n = 0 # y[0] = v0, y[1] = y[0]+dt/m*F(1), ...
for xv in x: # xv = x[n]
    if xv <= 0: y.append(v0) # n = 0
    else: y.append(y[n-1] + dt/m*F(xv)) # n = 1
    n += 1
```

```python
plt.plot(x, y)
plt.grid(True)
```

```python
def exv(t):
    if t <= 0: return v0 # t = 0
    elif t <= a: return a/m*(1-exp(-t))+ sin(t)/m + v0
    else: a/m*(1-exp(-a))+ sin(a)/m + v0 # t = a
```

```python
x2 = list(np.arange(-5, a + 5, 0.01))
y2 = []
for xv in x2:
    y2.append(exv(xv))
```

```python
plt.plot(x2, y2, 'r-')
plt.plot(x, y, 'b-')
plt.grid(True)
```
