```python
import matplotlib.pyplot as plt
import numpy as np
from math import *
```

```python
def F(t):
    a = 12
    if t >= 0 and t <= a: return a*exp(-t) + cos(t)
    else: return 0
```

```python
F(20)
```

```python
x=list(np.arange(-5, 20, 0.01))
y=[]
for xv in x:
    y.append(F(xv))
```

```python
plt.plot(x, y)
plt.grid(True)
```
