```python
import matplotlib.pyplot as plt
import numpy as np
from math import *
```

```python
a = 50 # 학번뒤 두자리 숫자
def myFun(x):
    return a*exp(x) + x*x - sin(x)
```

```python
y = myFun(5)
print(y)
```

```python
h = 1e-5
def diffFun(x):
    return (myFun(x+h)-myFun(x))/h
```

```python
x = list(np.arange(-5., 5., 0.001))
yp = [] #y의 수치 미분
wp = [] #y의 정확한 미분

for xv in x:
    yp.append(diffFun(xv))
    wp.append(a*exp(xv)+2*xv-cos(xv))
```

```python
plt.plot(x, yp, 'r-') # r: 빨간색, -: 실선
plt.plot(x, wp, 'b--') # b: 파란색, --: 점ㅅ
plt.grid(True)
```
